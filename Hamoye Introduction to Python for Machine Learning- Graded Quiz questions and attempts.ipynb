```python
#Question 1
import pandas as pd

dataset = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin1')

# Select only numeric columns for correlation (excluding 'Element Code')
numeric_dataset = dataset.select_dtypes(include='number').drop('Element Code', axis=1)

# Calculate the correlation of 'Element Code' with each year
correlations = numeric_dataset.corrwith(dataset['Element Code'])

# Find the year with the least correlation
least_correlated_year = correlations.idxmin()
least_correlation_value = correlations.min()

least_correlated_year, least_correlation_value
```

```python
#Question 3
#What is the total Protein supply quantity in Madagascar in 2015?
import pandas as pd

dataset = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin1')

# Filter the dataset for Madagascar and the year 2015
madagascar_data = dataset[(dataset['Area'] == 'Madagascar') & (dataset['Element'] == 'Protein supply quantity (g/capita/day)')]

# Calculate the total Protein supply quantity in Madagascar in 2015
protein_supply_2015 = madagascar_data['Y2015'].sum()

protein_supply_2015
```

```python
#Question 4
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?
import pandas as pd

dataset = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin1')
#Filtering to display just Area and Y2017 columns
Grp2017 = dataset[['Area', 'Y2017']]
#Summing the values with groupby() function 
Table= Grp2017.groupby(['Area']).sum()
#Sorting the values to be displayed by y2017 in descending order
Table = Table.sort_values(by='Y2017')
Table
```

```python
#Question 13
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?
import pandas as pd

dataset = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin1')

#Filtering the dataset to display just two columns of Area and Y2017
Area_Pop = dataset[['Area', 'Y2017']]
#grouping by Area column
Area_Val = Area_Pop.groupby('Area').sum()
#sorting the values to display in descending order i.e highest to lowest
Area_Val = Area_Val.sort_values(by='Y2017', ascending=False)
#Displaying the resulting Areas column sum in descending order
Area_Val
```

```python
#Question 19
#Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?
import pandas as pd
import statistics

dataset = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding='latin1')

Variation = dataset[['Element', 'Item', 'Unit', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']]

# Calculate the variance of the dataset
variance_2014 = statistics.variance(dataset['Y2014'].dropna())  # Drop NaN values before calculating variance
variance_2015 = statistics.variance(dataset['Y2015'].dropna())
variance_2016 = statistics.variance(dataset['Y2016'].dropna())
variance_2017 = statistics.variance(dataset['Y2017'].dropna())
variance_2018 = statistics.variance(dataset['Y2018'].dropna())

print('Year 2014:', variance_2014)
print('Year 2015:', variance_2015)
print('Year 2016:', variance_2016)
print('Year 2017:', variance_2017)
print('Year 2018:', variance_2018)
```
